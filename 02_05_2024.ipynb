```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread(r"pato.jpeg", 0)

roberts_x = cv2.filter2D(img, -1, np.array([[1, 0], [0, -1]]))
roberts_y = cv2.filter2D(img, -1, np.array([[0, 1], [-1, 0]]))

prewitt_x = cv2.filter2D(img, -1, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
prewitt_y = cv2.filter2D(img, -1, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

sobel_x = cv2.filter2D(img, -1, np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]]))
sobel_y = cv2.filter2D(img, -1, np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]]))


canny_x = cv2.Canny(img, cv2.CV_64F, 1, 0)
canny_y = cv2.Canny(img, cv2.CV_64F, 0, 1)

plt.figure(figsize=(12, 8))

plt.subplot(3, 5, 1)
plt.title("Original")
plt.imshow(img, cmap='gray')

plt.subplot(3, 5, 2)
plt.title("Roberts X")
plt.imshow(roberts_x, cmap='gray')

plt.subplot(3, 5, 3)
plt.title("Roberts Y")
plt.imshow(roberts_y, cmap='gray')

plt.subplot(3, 5, 4)
plt.title("Roberts Combined")
plt.imshow(np.sqrt(roberts_x**2 + roberts_y**2), cmap='gray')

plt.subplot(3, 5, 5)
plt.title("Prewitt X")
plt.imshow(prewitt_x, cmap='gray')

plt.subplot(3, 5, 6)
plt.title("Prewitt Y")
plt.imshow(prewitt_y, cmap='gray')

plt.subplot(3, 5, 7)
plt.title("Prewitt Combined")
plt.imshow(np.sqrt(prewitt_x**2 + prewitt_y**2), cmap='gray')

plt.subplot(3, 5, 8)
plt.title("Sobel X")
plt.imshow(sobel_x, cmap='gray')

plt.subplot(3, 5, 9)
plt.title("Sobel Y")
plt.imshow(sobel_y, cmap='gray')

plt.subplot(3, 5, 10)
plt.title("Sobel Combined")
plt.imshow(np.sqrt(sobel_x**2 + sobel_y**2), cmap='gray')

plt.subplot(3, 5, 11)
plt.title("Canny x")
plt.imshow(canny_x, cmap='gray')

plt.subplot(3, 5, 12)
plt.title("Canny y")
plt.imshow(canny_y, cmap='gray')

plt.subplot(3, 5, 13)
plt.title("Canny Combined")
plt.imshow(np.sqrt(canny_x**2 + canny_y**2), cmap='gray')

plt.show()
```
