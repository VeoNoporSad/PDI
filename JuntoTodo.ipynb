```python
import cv2
import matplotlib.pyplot as plt

imagen = cv2.imread(r"C:/Universidad/QuintoSemestre/PDI/Imagenes/dd5.jpg", 0) 

# Umbralizacin simple
_, umbralbi = cv2.threshold(imagen, 127, 255, cv2.THRESH_BINARY) # BLANCO Y NEGRO
_, umbral_inc_bi = cv2.threshold(imagen, 127, 255, cv2.THRESH_BINARY_INV) # Inversa Blanco y negro

# Umbralizacin adaptativa
umb_adap = cv2.adaptiveThreshold(imagen, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, blockSize=11, C=2)
umb_adap_gauss = cv2.adaptiveThreshold(imagen, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, blockSize=11, C=2)

plt.figure(figsize=(15, 10))

plt.subplot(231)
plt.imshow(imagen, cmap='gray')
plt.title("Imagen inicial")

plt.subplot(232)
plt.imshow(umbralbi, cmap="gray")
plt.title("Umbralizacin Simple")

plt.subplot(233)
plt.imshow(umbral_inc_bi, cmap="gray")
plt.title("Umbralizacin Simple Inversa")

plt.subplot(234)
plt.imshow(umbralbi, cmap="gray")
plt.title("Umbralizacin Simple")
plt.xlabel("Columna")
plt.ylabel("Fila")

plt.subplot(235)
plt.imshow(umb_adap, cmap="gray")
plt.title("Umbralizacin Adaptativa")
plt.xlabel("Columna")
plt.ylabel("Fila")

plt.subplot(236)
plt.imshow(umb_adap_gauss, cmap="gray")
plt.title("Umbralizacin Adaptativa Gaussiana")
plt.xlabel("Columna")
plt.ylabel("Fila")

# Aadir el histograma
plt.figure(figsize=(15, 3))
histo = cv2.calcHist([imagen], [0], None, [256], [0,256])
plt.plot(histo, color="black")
plt.title("Histograma")
plt.xlabel("Intensidad de pixel")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

```python
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import numpy as np

def get_gradient_2d(start, stop, width, height, is_horizontal):
    if is_horizontal:
        return np.tile(np.linspace(start, stop, width), (height, 1))
    else:
        return np.tile(np.linspace(start, stop, height), (width, 1)).T
    
def get_gradient_3d(width, height, start_list, stop_list, is_horizontal_list):
    result = np.zeros((height, width, len(start_list)), dtype=float)

    for i, (start, stop, is_horizontal) in enumerate(zip(start_list, stop_list, is_horizontal_list)):
        result[:, :, i] = get_gradient_2d(start, stop, width, height, is_horizontal)

    return result



array = get_gradient_3d(512, 256, (0, 0, 0), (255, 255, 255), (True, True, True))

imagee = Image.fromarray(np.uint8(array))


# Aplicando los umbrales
_, umb_trunc = cv2.threshold(array,127,200, cv2.THRESH_TRUNC)
_, umb_tozero = cv2.threshold(array,127,255, cv2.THRESH_TOZERO)
_, ro_INV = cv2.threshold(array,127,255,cv2.THRESH_TOZERO_INV)
_, Binario = cv2.threshold(array,127,255,cv2.THRESH_BINARY)
_, Binario_inv = cv2.threshold(array,127,255,cv2.THRESH_BINARY_INV)
_, GAUSSIAN = cv2.threshold(array,127,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C)
_, Mean = cv2.threshold(array,127,255,cv2.ADAPTIVE_THRESH_MEAN_C)





plt.figure(figsize=(10, 10))


plt.subplot(421)
plt.imshow(imagee, cmap='gray')
plt.title('Imagen original')

plt.subplot(422)
plt.imshow(umb_tozero, cmap='gray')
plt.title('Imagen Tozero')

plt.subplot(423)
plt.imshow(ro_INV, cmap='gray')
plt.title('Imagen Tozero inv')

plt.subplot(424)
plt.imshow(Binario, cmap='gray')
plt.title('Imagen Binaria')

plt.subplot(425)
plt.imshow(Binario_inv, cmap='gray')
plt.title('Imagen Binaria invertida')

plt.subplot(426)
plt.imshow(GAUSSIAN, cmap='gray')
plt.title('GAUSSIAN')

plt.subplot(427)
plt.imshow(Mean, cmap='gray')
plt.title('Mean')



plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def aplicar_umbral_binario(image):
    _, imagen_umbral = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
    plt.imshow(imagen_umbral, cmap='gray')
    plt.title('Umbral Binario')

def aplicar_umbral_invertido(image):
    _, imagen_umbral = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
    plt.imshow(imagen_umbral, cmap='gray')
    plt.title('Umbral Invertido')

def aplicar_umbral_truncado(image):
    _, imagen_umbral = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC)
    plt.imshow(imagen_umbral, cmap='gray')
    plt.title('Umbral Truncado')

def aplicar_umbral_a_cero(image):
    _, imagen_umbral = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)
    plt.imshow(imagen_umbral, cmap='gray')
    plt.title('Umbral a Cero')

def aplicar_umbral_a_cero_invertido(image):
    _, imagen_umbral = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)
    plt.imshow(imagen_umbral, cmap='gray')
    plt.title('Umbral a Cero Invertido')

def aplicar_umbral_dos_picos(image):
    _, imagen_umbral = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    plt.imshow(imagen_umbral, cmap='gray')
    plt.title('Umbral Dos Picos')

def aplicar_umbral_isodata(image):
    _, imagen_umbral = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    plt.imshow(imagen_umbral, cmap='gray')
    plt.title('Umbral Isodata')

def aplicar_umbral_adaptativo_mean(image):
    imagen_umbral = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
    plt.imshow(imagen_umbral, cmap='gray')
    plt.title('Umbral Adaptativo Media')

def aplicar_umbral_adaptativo_gaussiano(image):
    imagen_umbral = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
    plt.imshow(imagen_umbral, cmap='gray')
    plt.title('Umbral Adaptativo Gaussiano')

def umbralizacion_variable_local(image, block_size, threshold_offset):
    threshold_image = np.zeros_like(image)
    for y in range(0, image.shape[0], block_size):
        for x in range(0, image.shape[1], block_size):
            block = image[y:y+block_size, x:x+block_size]
            block_mean = np.mean(block)
            block_std = np.std(block)
            threshold = block_mean + (threshold_offset * block_std)
            threshold_image[y:y+block_size, x:x+block_size] = (block < threshold) * 255
    
    plt.imshow(threshold_image, cmap='gray')
    plt.title('Umbralizacin Variable Local')

def aplicar_umbral_OTSU(image):
    _, imagen_umbral = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    plt.imshow(imagen_umbral, cmap='gray')
    plt.title('Umbral OTSU')


def variable_local(image, block_size, threshold_offset):
    threshold_image = np.zeros_like(image)
    for y in range(0, image.shape[0], block_size):
        for x in range(0, image.shape[1], block_size):
            block = image[y:y+block_size, x:x+block_size]
            block_mean = np.mean(block)
            block_std = np.std(block)
            threshold = block_mean - (threshold_offset * block_std)
            threshold_image[y:y+block_size, x:x+block_size] = (block > threshold) * 255
    
    plt.imshow(threshold_image, cmap='gray')
    plt.title('Umbralizacin Variable por Bloques')

alto, ancho = 300, 300
imagen_degradada = np.zeros((alto, ancho), dtype=np.uint8)
inicio, fin = 0, 255
for i in range(alto):
    for j in range(ancho):
        imagen_degradada[i, j] = int(inicio + (fin - inicio) * (j / ancho))

plt.figure(figsize=(16, 12))

plt.subplot(4, 4, 1)
plt.imshow(imagen_degradada, cmap='gray')
plt.title('Original')

plt.subplot(4, 4, 2)
aplicar_umbral_binario(imagen_degradada)

plt.subplot(4, 4, 3)
aplicar_umbral_invertido(imagen_degradada)

plt.subplot(4, 4, 4)
aplicar_umbral_truncado(imagen_degradada)

plt.subplot(4, 4, 5)
aplicar_umbral_a_cero(imagen_degradada)

plt.subplot(4, 4, 6)
aplicar_umbral_a_cero_invertido(imagen_degradada)

plt.subplot(4, 4, 7)
aplicar_umbral_dos_picos(imagen_degradada)

plt.subplot(4, 4, 8)
aplicar_umbral_isodata(imagen_degradada)

plt.subplot(4, 4, 9)
aplicar_umbral_adaptativo_mean(imagen_degradada)

plt.subplot(4, 4, 10)
aplicar_umbral_adaptativo_gaussiano(imagen_degradada)

plt.subplot(4, 4, 11)
umbralizacion_variable_local(imagen_degradada, 25, 0.8)

plt.subplot(4, 4, 12)
variable_local(imagen_degradada, 25, 0.8)

plt.subplot(4, 4, 13)
aplicar_umbral_OTSU(imagen_degradada)

plt.tight_layout()
plt.show()
```
