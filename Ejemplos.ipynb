```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Cargar la imagen desde tu PC

original_image = cv2.imread(r'C:\Universidad\QuintoSemestre\PDI\Imagenes\grucara.jpeg')
# Convertir la imagen a escala de grises
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# Crear un degradado del mismo tamao que la imagen en escala de grises
height, width = gray_image.shape
gradient = np.zeros_like(gray_image)

# Aplicar un degradado a la imagen en escala de grises
for i in range(height):
    for j in range(width):
        gradient[i, j] = int(255 * (j / width))  # Valor de intensidad aumenta de izquierda a derecha

# Superponer el degradado sobre la imagen en escala de grises
gradient_on_gray = cv2.addWeighted(gray_image, 0.5, gradient, 0.5, 0)

# Mostrar la imagen original en escala de grises y la imagen con degradado superpuesto
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Imagen original en escala de grises')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gradient_on_gray, cmap='gray')
plt.title('Degradado superpuesto')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Cargar la imagen desde tu PC
original_image = cv2.imread(r'C:\Universidad\QuintoSemestre\PDI\Imagenes\grucara.jpeg')

# Convertir la imagen a escala de grises
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# Crear un degradado del mismo tamao que la imagen en escala de grises
height, width = gray_image.shape
gradient = np.zeros_like(gray_image)

# Aplicar un degradado a la imagen en escala de grises (de blanco a negro)
for i in range(height):
    for j in range(width):
        gradient[i, j] = 255 - int(255 * (i / height))  # Valor de intensidad disminuye de blanco a negro

# Superponer el degradado sobre la imagen en escala de grises
gradient_on_gray = cv2.addWeighted(gray_image, 0.5, gradient, 0.5, 0)

# Mostrar la imagen original en escala de grises y la imagen con degradado superpuesto
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Imagen original en escala de grises')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gradient_on_gray, cmap='gray')
plt.title('Degradado superpuesto (de blanco a negro)')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tamao de cada casilla del tablero de ajedrez
square_size = 100

# Tamao del tablero de ajedrez
board_size = 8 * square_size

# Crear una matriz para el tablero de ajedrez
board = np.ones((board_size, board_size), dtype=np.uint8) * 255

# Rellenar las casillas con colores alternados
for i in range(0, board_size, square_size):
    for j in range(0, board_size, square_size):
        if (i + j) % (2 * square_size) == 0:
            board[i:i + square_size, j:j + square_size] = 0

# Mostrar el tablero de ajedrez
plt.imshow(board, cmap='gray')
plt.axis('off')
plt.show()
```
