```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('elementos.jpg')

gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
_, binarizada = cv2.threshold(gris, 128, 255, 9)

borde,_ = cv2.findContours(binarizada, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cont = len(borde)

plt.figure(figsize=(20, 8))
plt.subplot(131)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(132)
plt.imshow(cv2.cvtColor(binarizada, cv2.COLOR_BGR2RGB))
plt.title(f'Contados:{cont}')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


imagen_original = cv2.imread('cristo.jpg')
imagen_original = cv2.cvtColor(imagen_original, cv2.COLOR_BGR2RGB)

imagen_empañada = cv2.GaussianBlur(imagen_original, (41, 41), 0)

img_fin = (0.3 * imagen_original + 0.7 * imagen_empañada).astype(np.uint8)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(imagen_original)

plt.subplot(1, 2, 2)
plt.title('Imagen Empañada (Convolución)')
plt.imshow(img_fin)

plt.show()
```
