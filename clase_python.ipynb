```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r"C:/Universidad/QuintoSemestre/PDI/Imagenes/dd3.jpg", cv2.IMREAD_GRAYSCALE)

#guardamos profundidad para mostrar
original_img = image.dtype

print('profundidad original de color:', original_img)

#Transformacion de imagen a bits
image_16bits = np.uint16(image)
nueva_img = image_16bits.dtype

print('profundidad original de color:', nueva_img)

#Guardamos la imagen tratada
cv2.imwrite('image_16bits.jpg', image_16bits)


img_guardada = cv2.imread('image_16bits.jpg', cv2.IMREAD_UNCHANGED)

plt.figure(figsize=(10,5))

plt.subplot(121)
plt.title("Original")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(122)
plt.title("Imagen Tratada 16 bits")
plt.imshow(img_guardada, cmap='gray')
plt.axis('off')

plt.show()
```
