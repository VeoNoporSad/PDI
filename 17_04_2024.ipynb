```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Cargar la imagen con aliasing
img = cv2.imread(r"Aliasing.jpg")

# Aplicar un filtro de suavizado para reducir el aliasing
img_suavizada = cv2.GaussianBlur(img, (5, 5), 0)

# Mostrar la imagen original y la imagen suavizada
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')

plt.subplot(122)
plt.imshow(cv2.cvtColor(img_suavizada, cv2.COLOR_BGR2RGB))
plt.title('Imagen Suavizada')

plt.show()
```

```python
import numpy as np
from scipy.fftpack import *
import imageio
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
plt.rcParams['figure.dpi'] = 150
```

```python
orig = imageio.imread(r'images/cat_original.jpg')
noisy = imageio.imread(r'images/cat_horizontal.jpg')
```

```python
orig_gray = np.dot(orig[...,:3], [0.2989, 0.5870, 0.1140])
noisy_gray = np.dot(noisy[...,:3], [0.2989, 0.5870, 0.1140])

plt.subplot(121)
plt.imshow(noisy_gray, cmap='gray')
plt.subplot(122)
plt.imshow(orig_gray, cmap='gray')
```
