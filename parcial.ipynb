```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Cargar la imagen desde tu PC
original_image = cv2.imread(r'C:\Universidad\QuintoSemestre\PDI\Imagenes\imagen1.jpg')

# Convertir la imagen a escala de grises
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

# Crear un degradado del mismo tamaño que la imagen en escala de grises
height, width = gray_image.shape
gradient = np.zeros_like(gray_image)

# Aplicar un degradado a la imagen en escala de grises (de blanco a negro)
for i in range(height):
    for j in range(width):
        gradient[i, j] = 255 - int(255 * (i / height))  # Valor de intensidad disminuye de blanco a negro

# Superponer el degradado sobre la imagen en escala de grises
gradient_on_gray = cv2.addWeighted(gray_image, 0.5, gradient, 0.5, 0)

# Mostrar la imagen original en escala de grises y la imagen con degradado superpuesto
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Imagen original en escala de grises')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gradient_on_gray, cmap='gray')
plt.title('Degradado superpuesto')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Cargar la imagen desde tu PC
original_image = cv2.imread(r'C:\Universidad\QuintoSemestre\PDI\Imagenes\parcial1.jpg')
celula = plt.imread(r'C:\Universidad\QuintoSemestre\PDI\Imagenes\parcial1.jpg')
color = ('r', 'g', 'b')
for canal, col in enumerate(color):

    histograma = cv2.calcHist([celula*255],[canal],None,[256], [0,256])
    plt.title("Histograma")
    plt.plot(histograma, color = col)
    plt.xlim([0,256])
plt.show()

def isodata(image):
    # Umbral inicial decidimos del espectro de 0 a 255
    threshold = 128
    
    while True:
        # Se divide la imagen en dos grupos según el umbral
        
        groupo1 = image[image <= threshold] # grupo donde el umbral es menor o igual al determinado
        groupo2 = image[image > threshold] # grupo donde el umbral es mayor al determinado
        
        # Se calcula la media de cada grupo
        media1 = np.mean(groupo1)
        media2 = np.mean(groupo2)
        
        # Nuevo umbral
        nuevo_umbral = (media1 + media2) / 2 # segun formula 
        
        # Si el nuevo umbral es igual al umbral anterior, se detiene el bucle
        if abs(nuevo_umbral - threshold) < 0.5:
            break
        
        threshold = nuevo_umbral
    
    return threshold

def solarizar(image,umbral):
    img_solarizada = np.where(image<umbral,image,255-image)
    return img_solarizada

def posterizar(image,nivel):
    f_nvl=255/nivel
    img_poster=np.round(image/f_nvl)*f_nvl
    return img_poster.astype(np.uint8)

img = cv2.imread(r"C:/Universidad/QuintoSemestre/PDI/Imagenes/parcial1.jpg", cv2.COLOR_BGR2RGB)

parte = img[30:80,100:200,:]
plt.imsave("Parte.jpg", parte)
img = cv2.imread(r"Parte.jpg")
plt.show()

img_gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
umbral_isodata = isodata(img)
_, imagen_umbralizada = cv2.threshold(img, umbral_isodata, 255, cv2.THRESH_BINARY) #OJO el "_" es una variable pero que ignoramos (almacena el valor del umbral) 
##############################
umbral = 127
img_sol = solarizar(img_gris, umbral)
##############################
nivel = 4
img_posterizada = posterizar(img_gris, nivel)
##############################
plt.subplot(4,1,1)
plt.imshow(img_gris)
plt.title("Original")

plt.subplot(4,1,2)
plt.imshow(img_sol)
plt.title("Solarizada")

plt.subplot(4,1,3)
plt.imshow(img_posterizada)
plt.title("Posterizada")

plt.subplot(4,1,4)
plt.imshow(imagen_umbralizada)
plt.title("Umbralizada Isodata")

plt.tight_layout()
plt.show()
plt.imsave("Solarizada.jpg", img_sol)
plt.imsave("img_posterizada.jpg", img_posterizada)
plt.imsave("imagen_umbralizada.jpg", imagen_umbralizada)
```
