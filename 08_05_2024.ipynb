```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def etiquetar(image):
    _,image_bin = cv2.threshold(image,127,255,cv2.THRESH_BINARY)

    contorno,_ = cv2.findContours(image_bin, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    img_et = np.zeros_like(image, dtype=np.uint8)

    for i, contorno1 in enumerate(contorno):
        cv2.drawContours(img_et, [contorno1], -1, (i+1), -1)
    return img_et

img_original = cv2.imread(r'montapuercos.jpg', cv2.IMREAD_GRAYSCALE)
imagen_etiquetada = etiquetar(img_original)


plt.figure(figsize=(15, 8))
plt.subplot(121)
plt.title('Original')
plt.imshow(img_original ,cmap='gray')
plt.axis('off')

plt.subplot(122)
plt.title('Etiquetada')
plt.imshow(imagen_etiquetada ,cmap='gray')
plt.axis('off')
```

```python
import cv2
import skimage.measure as p
import matplotlib.pyplot as plt
import numpy as np

# Lee la imagen
img = cv2.imread(r'montapuercos.jpg', cv2.IMREAD_GRAYSCALE)

# Aplica umbralizacin binaria
_, binaria = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Etiqueta las regiones conectadas
img_etiquetada = p.label(binaria, connectivity=1)

# Muestra las imgenes
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Imagen original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Imagen etiquetada")
plt.imshow(img_etiquetada, cmap='jet')
plt.axis('off')

plt.show()
```
