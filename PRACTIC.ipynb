dilatacion binaria

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('cancer.jpg', cv2.IMREAD_GRAYSCALE)
_, img_bin = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
mascara = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
img_fin = cv2.dilate(img_bin, mascara)
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_bin, cmap='gray')
plt.title('Binary Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_fin, cmap='gray')
plt.title('Dilated Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

close

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('cancer.jpg', cv2.IMREAD_GRAYSCALE)
mascara = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (9, 9))
cerradura = cv2.morphologyEx(img, cv2.MORPH_CLOSE, mascara)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cerradura, cmap='gray')
plt.title('Image with Closing')
plt.axis('off')

plt.tight_layout()
plt.show()
```

recontrucciones de apertura

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('moto.jpg', cv2.IMREAD_GRAYSCALE)
mascara = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
img_apertura = cv2.morphologyEx(img, cv2.MORPH_OPEN, mascara)
k_reconStruccion = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
img_fin = cv2.morphologyEx(img_apertura, cv2.MORPH_CLOSE, k_reconStruccion)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_apertura, cmap='gray')
plt.title('Image de apertura')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_fin, cmap='gray')
plt.title('Image reconstrucciones de apertura')
plt.axis('off')

plt.tight_layout()
plt.show()
```

recontrucciones de cerradura

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('elementos.jpg', cv2.IMREAD_GRAYSCALE)
mascara = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
img_cierre = cv2.morphologyEx(img, cv2.MORPH_OPEN, mascara)
k_reconStruccion = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
img_fin = cv2.morphologyEx(img_cierre, cv2.MORPH_OPEN, k_reconStruccion)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_cierre, cmap='gray')
plt.title('Image de cierre')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_fin, cmap='gray')
plt.title('Image reconstrucciones de cierre')
plt.axis('off')

plt.tight_layout()
plt.show()
```
