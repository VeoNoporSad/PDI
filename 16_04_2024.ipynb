Convolucion 
Es el tratamiento de otra matrz por otro llamado Kernel

La matriz de de convolucion es de 3x3

Procesamiento en el Dominio Espacial

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# Definir la imagen de ejemplo
image = np.array([
    [100, 120, 130, 110],
    [90, 80, 70, 95],
    [75, 85, 95, 100],
    [110, 100, 120, 130]
])

# Definir el kernel de suavizado
kernel = np.array([
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9]
])

# Realizar la convoluci√≥n
convolved_image = convolve2d(image, kernel, mode='same', boundary='fill', fillvalue=0)

# Mostrar la imagen original y la imagen convolucionada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagen original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(convolved_image, cmap='gray')
plt.title('Imagen convolucionada')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread(r"MontaPuercos.jpg", 0)
mascara  = np.array([[ 0, -1,  0], [-1, 5, -1], [0, -1, 0]])
mascara2 = np.array([[1, 1, 1],[1, 1, 1],[1, 1, 1]]) / 9
mascara3 = np.array([[0, 1, 0], [ 1, -4,  1], [0, 1, 0]])

imgFiltro1 = cv2.filter2D(img, -1, mascara)
imgFiltro2 = cv2.filter2D(img, -1, mascara2)
imgFiltro3 = cv2.filter2D(img, -1, mascara3)

plt.figure(figsize=(20, 5))
plt.subplot(141)
plt.title('Original')
plt.imshow(img ,cmap='gray')
plt.axis('off')

plt.subplot(142)
plt.title('Filtro 1')
plt.imshow(imgFiltro1 ,cmap='gray')
plt.axis('off')

plt.subplot(143)
plt.title('Filtro 2')
plt.imshow(imgFiltro2 ,cmap='gray')
plt.axis('off')

plt.subplot(144)
plt.title('Filtro 3')
plt.imshow(imgFiltro3 ,cmap='gray')
plt.axis('off')

plt.show()
```
